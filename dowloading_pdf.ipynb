```python
import pdfplumber

pdf_filename = 'Bihar_Govt_Calendar.pdf'

# Open the PDF file and extract text
all_text = ''
with pdfplumber.open(pdf_filename) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        all_text += text + '\n'

# Print the extracted text (optional, for checking)
print(all_text)
```

```python
import re

# Define a regex pattern to find festival dates
pattern = re.compile(r'(\d{1,2} [A-Za-z]+ - .+)')

# Find all matches in the extracted text
festival_dates = pattern.findall(all_text)

# Print the extracted festival dates
for date in festival_dates:
    print(date)
```

```python
import requests

pdf_url = 'https://www.biharcommercialtax.gov.in/bweb/download/general/Bihar%20Govt.%20Calendar.pdf'

# Download the PDF with SSL verification disabled
response = requests.get(pdf_url, verify=False)

pdf_filename = 'Bihar_Govt_Calendar.pdf'
with open(pdf_filename, 'wb') as file:
    file.write(response.content)

print(f'PDF downloaded and saved as {pdf_filename}')
```
